```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
file_men = '/u/tlooby/results/eqtools_v_menard/q_204118_00850_menard.dat'
file_eq = '/u/tlooby/results/eqtools_v_menard/q_204118_00850_eqtools.dat'
data_men = np.loadtxt(file_men, comments='#', delimiter='\t')
data_eq = np.loadtxt(file_eq, comments='#', delimiter='\t')
plt.plot(data_men[:,0], data_men[:,1], label='Menard gfile', marker='^')
plt.plot(data_eq[:,0], data_eq[:,1], label='Eqtools gfile')
plt.xlabel('S [m]')
plt.ylabel('q [MW/m^2]')
plt.legend(loc='upper right')
plt.title("EQTOOLS vs. Menard's g-file Creators \n \
          P=2.8MW; lq=5mm S=2.1mm ")
#plt.show()
```
